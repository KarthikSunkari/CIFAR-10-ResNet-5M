# CIFAR-10 Custom ResNet Model

## Overview
Custom ResNet model with Squeeze-and-Excitation (SE) blocks to classify images from the CIFAR-10 dataset. Below is a summary of the key steps:


---

## Steps

### Data Preparation
- Loads CIFAR-10 training and test data, reshapes images, and combines batches.
- Normalizes pixel values and applies mean and standard deviation normalization.
- One-hot encodes labels for multi-class classification.
- Splits training data into training and validation sets.

---

### Model Design
- Builds an 80-layer ResNet model with SE blocks for enhanced feature extraction.
- Includes residual connections and SE optimization for improved accuracy.

---

### Training
- Augments training data with random transformations (rotation, flipping, zooming).
- Implements learning rate scheduling and uses callbacks like early stopping, model checkpointing, and learning rate reduction.
- Trains the model using augmented data for 400 epochs.

---

### Evaluation and Prediction
- Applies test-time augmentation (TTA) to improve predictions on test data.
- Generates predictions and creates a submission file with class labels.

---

### Visualization
- Plots training and validation accuracy/loss curves to monitor performance.

---

## Output
Validation Accuracy (10% of Dataset) : 94.10%

```python
import numpy as np
import pandas as pd
import os
import pickle
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import LearningRateScheduler, ReduceLROnPlateau, ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, AveragePooling2D, Flatten, Dense, add
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split


# Load CIFAR-10 batch function
def load_cifar_batch(file):
    with open(file, 'rb') as fo:
        batch_dict = pickle.load(fo, encoding='bytes')
    return batch_dict

# Loading all training batches
cifar10_dir = '/scratch/ks7929/Project1/cifar-10-python/cifar-10-batches-py'
train_data = []
train_labels = []

for i in range(1, 6):
    batch_dict = load_cifar_batch(os.path.join(cifar10_dir, f'data_batch_{i}'))
    batch_data = batch_dict[b'data']
    batch_data = batch_data.reshape((10000, 3, 32, 32)).transpose(0, 2, 3, 1)
    train_data.append(batch_data)
    train_labels.extend(batch_dict[b'labels'])

X_train = np.concatenate(train_data)
y_train = np.array(train_labels)

# Loading test data
test_data = load_cifar_batch('/scratch/ks7929/Project1/cifar_test_nolabel.pkl')
X_test = test_data[b'data']


# Normalizing data
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0


# Applying mean and std normalization
mean = np.mean(X_train, axis=(0, 1, 2))
std = np.std(X_train, axis=(0, 1, 2))
X_train = (X_train - mean) / (std + 1e-7)
X_test = (X_test - mean) / (std + 1e-7)


# One-hot encode labels
y_train_one_hot = tf.keras.utils.to_categorical(y_train, 10)


# Splitting training data for validation
X_train, X_val, y_train_one_hot, y_val_one_hot = train_test_split(
    X_train, y_train_one_hot, test_size=0.1, random_state=42)


# Enhanced data augmentation
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    shear_range=0.1,
    zoom_range=0.1
)
datagen.fit(X_train)


# Learning Rate Scehduler
def lr_schedule(epoch):
    initial_lr = 0.001
    if epoch < 5:
        return initial_lr * ((epoch + 1) / 5)
    elif epoch < 20:
        return initial_lr
    elif epoch < 100:
        return initial_lr * 0.1
    elif epoch < 250:
        return initial_lr * 0.01
    else:
        return initial_lr * 0.001


# Define Squeeze-and-Excitation block
def squeeze_and_excitation_block(x, num_filters, ratio=16):
    """Squeeze-and-Excitation block."""
    squeeze = layers.GlobalAveragePooling2D()(x)
    squeeze = layers.Dense(num_filters // ratio, activation='relu')(squeeze)
    squeeze = layers.Dense(num_filters, activation='sigmoid')(squeeze)
    squeeze = layers.Reshape((1, 1, num_filters))(squeeze)
    return layers.Multiply()([x, squeeze])


# Residual block with SE optimization
def resnet_block_se(inputs, num_filters=32, strides=1):
    """Residual block with Squeeze-and-Excitation."""
    x = Conv2D(num_filters, kernel_size=3, strides=strides,
               padding='same', kernel_initializer='he_normal',
               kernel_regularizer=regularizers.l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(num_filters, kernel_size=3,
               padding='same', kernel_initializer='he_normal',
               kernel_regularizer=regularizers.l2(1e-4))(x)
    x = BatchNormalization()(x)


    # Apply SE block
    x = squeeze_and_excitation_block(x, num_filters)

    if strides != 1 or inputs.shape[-1] != num_filters:
        shortcut = Conv2D(num_filters,
                          kernel_size=1,
                          strides=strides,
                          padding='same',
                          kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(inputs)
        shortcut = BatchNormalization()(shortcut)
    else:
        shortcut = inputs

    x = add([x, shortcut])
    x = Activation('relu')(x)

    return x


# Custom ResNet model with SE blocks (80 layers)
def custom_resnet_se(input_shape=(32, 32, 3), num_classes=10):
    """Custom ResNet model with Squeeze-and-Excitation blocks."""
    inputs = Input(shape=input_shape)


    # Initial convolutional layer
    x = Conv2D(32,
               kernel_size=3,
               padding='same',
               kernel_initializer='he_normal',
               kernel_regularizer=regularizers.l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)


    # Residual blocks: [12 blocks per stage]
    num_blocks_per_stage = [12] * 3
    num_filters_per_stage = [32, 64, 128]


    for stage_idx in range(len(num_blocks_per_stage)):
        for block_idx in range(num_blocks_per_stage[stage_idx]):
            strides = 2 if block_idx == 0 and stage_idx > 0 else 1
            x = resnet_block_se(x,
                                num_filters=num_filters_per_stage[stage_idx],
                                strides=strides)


    # Final layers: Global average pooling and dense layer for classification
    x = AveragePooling2D(pool_size=8)(x)
    x = Flatten()(x)
    outputs = Dense(num_classes,
                    activation='softmax',
                    kernel_initializer='he_normal')(x)

    model = Model(inputs=inputs, outputs=outputs)

    return model


# Initialize the model (80 layers with filters starting at 32)
model = custom_resnet_se(input_shape=(32, 32, 3), num_classes=10)


model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


model.summary()


# Print the total number of parameters
total_params = model.count_params()
print(f'Total number of parameters: {total_params}')


# Callbacks for training
lr_scheduler = LearningRateScheduler(lr_schedule)
reduce_lr_on_plateau_callback = ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                                  patience=10,
                                                  verbose=True,
                                                  min_lr=1e-6)


checkpoint_callback = ModelCheckpoint('best_model_custom_res.keras',
                                       save_best_only=True,
                                       verbose=True)


early_stopping_callback = EarlyStopping(monitor='val_loss',
                                        patience=200,
                                        restore_best_weights=True)


# Train the model
history = model.fit(datagen.flow(X_train,
                                 y_train_one_hot,
                                 batch_size=128),
                    epochs=400,
                    validation_data=(X_val,
                                     y_val_one_hot),
                    callbacks=[lr_scheduler,
                               checkpoint_callback, reduce_lr_on_plateau_callback,
                               early_stopping_callback], verbose=1)


# Load the best model for evaluation on test data.
model=tf.keras.models.load_model('best_model_custom_res.keras')


# Test-time augmentation function
def tta_predict(model, images, n_augmentations=10):
    predictions = np.zeros((images.shape[0], 10))

    # Original images prediction
    predictions += model.predict(images)

    # Test-time augmentation generator
    tta_gen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        zoom_range=0.1
    )


    # Augmented predictions
    for i in range(n_augmentations - 1):
        aug_images = np.copy(images)
        for j in range(len(images)):
            aug_images[j] = tta_gen.random_transform(images[j])
        predictions += model.predict(aug_images)


    # Average predictions
    predictions /= n_augmentations
    return predictions


# Make predictions with test-time augmentation
predictions = tta_predict(model, X_test, n_augmentations=5)
predicted_classes = np.argmax(predictions, axis=1)


# Create submission file
submission_df = pd.DataFrame({'ID': range(len(predicted_classes)), 'Labels': predicted_classes})
submission_df.to_csv('submission_custom_res.csv', index=False)


# Plot training & validation accuracy values
plt.figure(figsize=(12, 6))


# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')


# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()


```
